```python
from matplotlib import pyplot as plt
import seaborn as snsxdd
import numpy as np
import pandas as pd
```

```python
df1 = sns.load_dataset('diamonds', engine='python', error_bad_lines=False, warn_bad_lines=False)
df1
df = pd.read_csv('diamonds.csv')
df_1 = df.drop(columns = ['Unnamed: 0','x','y','z'])
df_1
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"

#code here
f, ax = plt.subplots(figsize=(13, 13))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["IF", "VVS1", "VVS2", "VS1", "VS2", "SI1", "SI2", "I1"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                hue_order=clarity_ranking,
                palette = palette,
                sizes=(1, 8), linewidth=0,
                data=df, ax=ax)
plt.show()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here
with sns.axes_style(style='ticks'):
    g = sns.catplot(x='clarity', y='price', data=df, order=["IF", "VVS1", "VVS2", "VS1", "VS2", "SI1", "SI2", "I1"], col='cut', col_order=['Ideal','Premium','Very Good','Good','Fair'], kind='box', showfliers=False, height=4)
    g.set_axis_labels("Clarity", "Price");
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df_sample = df_1.sample(n=300)

#code here
sns.pairplot(df_sample, hue='cut', hue_order=["Ideal", "Premium", "Very Good", "Good", "Fair"], markers=['p','o','s','D','X'], corner=True);
```
