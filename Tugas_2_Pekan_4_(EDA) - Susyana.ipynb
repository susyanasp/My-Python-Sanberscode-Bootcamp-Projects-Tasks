Jelaskan perbedaan antara Univariate, Bivariate dan Multivariate Analysis 

Jawaban:

Univariate analysis: sebuah teknik dalam memahami dan melakukan eksplorasi data & analisa fitur data tunggal. Saat melakukan analisis ini, jenis-jenis grafik: histogram, bar chart, density plots, distribution plots, dan lainnya.

Bivariate analysis: salah satu cara untuk menggunakan koefisien korelasi dalam rangka menganalisis dua variabel dan menemukan apakah ada relasi antara dua variabel tersebut. Jenis-jenis grafik yang digunakan: scatter plot, mosaic plots, histogram/barplot, heatmaps, dan linechart.

Multivariate analysis: salah satu cara untuk menganalisis lebih dari dua variabel di waktu yang sama, trends yang dihasilkan bisa menjadi multidimensi secara alami, dan akan membantu untuk memahami manakah data yang memiliki tren yang kompleks pada kombinasi atribut. Maka, ini bisa menentukan manakah algoritma yang tepat yang bisa digunakan untuk prediktif analisis.

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
*   Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g
*   Jelaskan insight yang diperoleh dari ke dua grafik tersebut.





```python
#code here
sums = df.groupby('species')['flipper_length_mm'].mean()

fig = plt.figure(figsize =(10, 7))
plt.pie(sums, labels=sums.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()

fig = plt.figure(figsize =(10, 8))
plt.hist(df['body_mass_g'], bins=10)
plt.xlabel("Body_mass_g")
plt.ylabel("Count")
plt.show()
```

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
#code here
sns.set({'figure.figsize':(7,7)})

with sns.axes_style('white'):
    sns.catplot(x='sex', y='bill_length_mm', data=df, kind='bar')
plt.xlabel("sex")
plt.ylabel("bill_length_mm")
plt.show()

sns.set({'figure.figsize':(12,8)})

with sns.axes_style('white'):
    sns.scatterplot(x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm', data=df)
plt.show()
```
