Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

plt.plot(x, cos, label='Cosx', color='r')
plt.plot(x, cos3, label='3 Cosx', linestyle='dashed', color='g')
plt.plot(x, sin, label='Sin(x-90)', linestyle='dashdot', color='b')
plt.plot(x, sin2, label='2 Sinx', marker='x', linestyle='None', color='y')
plt.legend(loc='center')

plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Grafik Cos Sin')

plt.show()
```
